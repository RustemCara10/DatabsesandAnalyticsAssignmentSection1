Explain how to perform data manipulation and transformation tasks using built-in functions and
packages in R.
Can you demonstrate three examples of data manipulation tasks using R functions and packages?


Data manipulation in R typically involves processes like sorting, filtering, summarizing, and transforming data. This can be efficiently handled using packages like dplyr and tidyr, which are part of the tidyverse suite.

```r
#Example1 Filtering allows us to subset the data based on specific criteria. For instance,
# if we want to analyze buildings with high glazing area and specific orientations we do the following.

library(dplyr)
high_glaze_data <- data %>%
  filter(Glazing.Area > 0.3, Orientation %in% c(2, 3))
print(head(high_glaze_data))
```

```r
#Example 2 :we can add new variables that are functions of existing variables.
# For example, calculating the total insulated area as a sum of wall and roof areas.
buildings <- data %>%
  mutate(Insulated.Area = Wall.Area + Roof.Area)
print(head(buildings))
```

```r
# Example 3 : Summarizing data helps to understand distributions or aggregate metrics.
# For example, calculating the average Heating.Load and Cooling.Load for each Overall.Height.
summary_data <- data %>%
  group_by(Overall.Height) %>%
  summarise(
    Average_Heating_Load = mean(Heating.Load),
    Average_Cooling_Load = mean(Cooling.Load)
  )
print(head(summary_data))
```

Task 5 :Creating informative and visually appealing plots
using R.

To create informative and visually appealing plots i have use ggplot and i have given three examples using histogram, bar and line.

```r
# This R script creates a histogram to visualize the distribution of heating loads in my dataset.
ggplot(data, aes(x = `Heating.Load`)) +
  geom_histogram(binwidth = 1, fill = "blue", color = "black") +
  labs(title = "Distribution of Heating Load", x = "Heating Load", y = "Frequency")
```

```r
# This R script summarises the Average Cooling load, groups it by Overall Height
# and visualizes the Average Cooling Load by Overall Height .
height_cooling <- data %>%
  group_by(Overall.Height) %>%
  summarise(Average_Cooling = mean(Cooling.Load))

ggplot(height_cooling, aes(x = as.factor(Overall.Height), y = Average_Cooling)) +
  geom_bar(stat = "identity", fill = "steelblue") +
  labs(title = "Average Cooling Load by Overall Height",
       x = "Overall Height",
       y = "Average Cooling Load")
```

```r
#This line plot summarises the Average Cooling load, groups it by
# glazing area distribution and visualizes the effect of Glazing Area Distribution on Cooling Load.
glazing_effect <- data %>%
  group_by(Glazing.Area.Distribution) %>%
  summarise(Average_Cooling = mean(Cooling.Load))

ggplot(glazing_effect, aes(x = Glazing.Area.Distribution, y = Average_Cooling)) +
  geom_line() +
  labs(title = "Effect of Glazing Area Distribution on Cooling Load",
       x = "Glazing Area Distribution",
       y = "Average Cooling Load")
```
